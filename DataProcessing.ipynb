```python
import pandas as pd
import json
```

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)

df_match= pd.DataFrame(data[0]["matchSummary"])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'},axis=1,inplace=True)
df_match.head()

df_match.to_csv('t20_csv_files/matchSummary1.csv',index=False)
```

```python
match_id_dict= {}

for index,row in df_match.iterrows():
    key1= row["team1"] + " Vs " + row["team2"]
    key2= row["team2"] + " Vs " + row["team1"]

    match_id_dict[key1]= row["match_id"]
    match_id_dict[key2]=row["match_id"]

# match_id_dict
```

**Batting Summary**

```python
with open('t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)

data

all_records = []
for rec in data:
    all_records.extend(rec["battingSummary"])

```

```python
df_batting = pd.DataFrame(all_records)
df_batting.head(10)
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
df_batting.head()
```

```python
df_batting.drop(columns=["dismissal"],inplace=True)
```

```python
df_batting.tail(20)
```

```python
df_batting["batsmanName"]= df_batting['batsmanName'].apply(lambda x:x.replace('â€',''))
df_batting.head(20)
```

```python
df_batting["match_id"] = df_batting["match"].map(match_id_dict)
```

```python
df_batting.head(30)
```

```python
df_batting.shape 
```

```python
df_batting.to_csv('t20_csv_files/fact_batting_summary.csv',index=False)
```

**Bowling Summary**

```python
with open('t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)

# data

```

```python
data_list = []

for rec in data:
    data_list.extend(rec['bowlingSummary'])

# data_list
```

```python
df_bowlling = pd.DataFrame(data_list)
df_bowlling.head(10)
```

```python
df_bowlling["match_id"] = df_bowlling["match"].map(match_id_dict)
df_bowlling["bowlerName"] = df_bowlling["bowlerName"].apply(lambda x:x.replace("â€",""))
df_bowlling["bowlerName"]= df_bowlling["bowlerName"].apply(lambda x:x.replace("â€\xa0",""))
df_bowlling.tail()
```

```python
df_bowlling.to_csv("t20_csv_files/fact_bowling_summary.csv",index=False)
```
